```python
mean = 0
std = 2
plt.figure(figsize = (15, 8))
x = np.linspace(mean - 3*std, mean + 3*std, 100)
plt.plot(x, scipy.stats.norm.pdf(x, mean, std))
plt.axvline(x = mean - std, c = 'blue')
plt.axvline(x = mean + std, c = 'blue')
plt.axvline(x = mean - 2*std, c = 'red')
plt.axvline(x = mean + 2*std, c = 'red')
plt.axvline(x = mean - 3*std, c = 'black')
plt.axvline(x = mean + 3*std, c = 'black')
```

```python
plt.figure(figsize = (10, 5))
sn.boxplot(trends_df["Age"])
plt.axhline(trends_df["Age"].describe()['25%'], color = 'red', label = 'Q1')
plt.axhline(trends_df["Age"].describe()['50%'], color = 'yellow', label = 'Q2')
plt.axhline(trends_df["Age"].describe()['75%'], color = 'blue', label = 'Q3')

IQR = trends_df["Age"].describe()['75%'] - trends_df["Age"].trends_df()['25%']

plt.annotate('Batas Atas', (trends_df["Age"].describe()['75%'] + 1.5*IQR, 0.2),
             xytext = (trends_df["Age"].describe()['75%'] + 1.5*IQR, 0.4),
             arrowprops = dict(facecolor = 'blue'), fontsize = 13 )

plt.annotate('Batas Bawah', (trends_df["Age"].describe()['25%'] - 1.5*IQR, 0.2),
             xytext = (trends_df["Age"].describe()['25%'] - 1.5*IQR, 0.4),
             arrowprops = dict(facecolor = 'blue'), fontsize = 13 )
plt.legend()
```
